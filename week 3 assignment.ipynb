# For this problem, you will be using the data set us pop data.csv. This data is from taken from the US Census has the US population every ten years from 1790 to 2010.

### a) Import the data and create two new columns. Create one column that is the number of years since 1790. Create another column that is the population in millions.

```python
#importing neccessary libraries
import numpy as np
import pandas as pd

#importing data
df = pd.read_csv(r"C:\Users\bharo\OneDrive\Desktop\math for data science\week3data\us_pop_data.csv")
df['number_years'] = df.year - 1790 # new column with no. of years since 1790
df['us_pop in mil'] = df.us_pop /1000000# new column with population in millions
df
```

## (b) Plot the US population (in millions) versus the years since 1790.

```python
import matplotlib.pyplot as plt

x = df['number_years']
y = df['us_pop in mil'] 


# ploting us population Vs years since 1790
plt.xlabel(' years since 1790', color = 'g')
plt.ylabel('US population (in millions)', color = 'g')
plt.title("US population (in millions) versus the years since 1790", color = 'g')
plt.plot(x,y)
plt.show()
```

## (c) Create a linear regression model to predict the US population (in millions) t years from 1790. Find and report the R2-value of this model

```python
#ploting the above data
plt.xlabel(' years since 1790', color = 'g')
plt.ylabel('US population (in millions)', color = 'g')
plt.title("US population (in millions) versus the years since 1790", color = 'g')
plt.plot(x,y)

#using sklearn for linear Regression line
from sklearn.linear_model import LinearRegression
x1 = np.array(x).reshape(-1,1) # converting x elements to array and reshaping it to two dimention
lr = LinearRegression()
lr.fit(x1,y) # fitting x1, y to the Linear Regression function
y1 = lr.predict(x1) # predicts y = mx+b
plt.plot(x1,y1)

# importing r2_scrore from sklearn.metrics for r2 value
from sklearn.metrics import r2_score
 
r2_value = r2_score(y,y1)# using r2_score()
print('R2 VALUE IS: ', r2_value)

```

##  (d) Create another new column in your data by squaring the number of years since 1790.

## e) Run another linear regression, where your input feature is the square of the number of years since 1790. Find and report the R2-value of this model.

```python
x2 = df['no of years^2'] # assigning 'no of years^2' column to x2 variable

plt.xlabel('no of years squared', color = 'g')
plt.ylabel('us_pop in mil', color = 'g')
plt.title('Number of years squared Vs US pop in millions')

x3 = np.array(x2).reshape(-1,1)
y3 = df['us_pop in mil']
lr = LinearRegression()
lr.fit(x3,y3)
plt.plot(x2,y3)
plt.plot(x3,lr.predict(x3))
plt.show()

from sklearn.metrics import r2_score
r2_value = r2_score(y3,lr.predict(x3))
print('R2 VALUE IS: ', r2_value)
```
